```
streaming_data = streaming_data.astype({"Age":"int"})
```

```
streaming_data = streaming_data.astype({"Time_on_platform":"int"})
streaming_data = streaming_data.astype({"Devices_connected":"int"})
streaming_data = streaming_data.astype({"Num_streaming_services":"int"})
streaming_data = streaming_data.astype({"Avg_rating":"int"})
streaming_data = streaming_data.astype({"Churned":"int"})
```

```
streaming_data.dtypes
```

```
streaming_data['Churned'] = streaming_data['Churned'].replace({0: 'No', 1: 'Yes'})
```

```
streaming_data
```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

le.fit(streaming_data.Gender)
streaming_data.Gender = le.transform(streaming_data.Gender)

le.fit(streaming_data.Subscription_type)
streaming_data.Subscription_type = le.transform(streaming_data.Subscription_type)
```

```
X = streaming_data[["Age","Gender","Time_on_platform","Devices_connected","Subscription_type","Num_streaming_services","Num_active_profiles","Avg_rating"]]


y = streaming_data[["Churned"]].copy()
y = pd.DataFrame(y)
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

```
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
lr = model.fit(X_train, y_train)
```

```
Y_pred = lr.predict(X_test)
```

```
lr.predict(X_test)
```

```
X_test.assign(churn = y_test).assign(label = lr.predict(X_test))
```

```
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```
cm = confusion_matrix(y_test, Y_pred)
```

```
import matplotlib.pyplot as plt
```

```
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=lr.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.title("Matriz de Confusão")
plt.show()
```

```
from sklearn.metrics import accuracy_score, balanced_accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```
print(f"Acurácia (Treino): {accuracy_score(y_train, lr.predict(X_train))}")
print(f"Acurácia (Teste): {accuracy_score(y_test, lr.predict(X_test))}")
print("====================================")
print(f"Acurácia Balanceada (Treino): {balanced_accuracy_score(y_train, lr.predict(X_train))}")
print(f"Acurácia Balanceada (Teste): {balanced_accuracy_score(y_test, lr.predict(X_test))}")
print("====================================")
print(f"Precision (Treino): {precision_score(y_train, lr.predict(X_train))}")
print(f"Precision (Teste): {precision_score(y_test, lr.predict(X_test))}")
print("====================================")
print(f"Recall (Treino): {recall_score(y_train, lr.predict(X_train))}")
print(f"Recall (Teste): {recall_score(y_test, lr.predict(X_test))}")
print("====================================")
print(f"F1-Score (Treino): {f1_score(y_train, lr.predict(X_train))}")
print(f"F1-Score (Teste): {f1_score(y_test, lr.predict(X_test))}")
print("====================================")
print(f"ROCAUC (Treino): {roc_auc_score(y_train, lr.predict_proba(X_train)[:,1])}")
print(f"ROCAUC (Teste): {roc_auc_score(y_test, lr.predict_proba(X_test)[:,1])}")
```
